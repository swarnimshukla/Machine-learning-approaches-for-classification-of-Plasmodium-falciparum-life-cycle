```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

data1 = [1.28, 1.19, 1.14]
data2 = [1.28, 1.25, 1.26]
r4 = [1, 2, 3]
width =0.2
fig = plt.subplots(figsize =(12, 8))
br1 = np.arange(len(data1)) 
br2 = [x + width for x in br1]

 
plt.bar(br1, data1, width=width, label = "Classification without feature selection")
plt.bar(br2, data2, width=width, label = "Classification with feature selection", color = 'red')
plt.legend(loc="upper left",  bbox_to_anchor = (1.05, 0.6))
plt.xticks([r + width - 0.16 for r in range(len(data1))], ['SVM', 'LR', 'RF'], fontsize = 12)
plt.ylabel("Mutual Information in bits",  fontsize=18)
label1 = ['1.28', '1.19', '1.14']
label2 = ['1.28', '1.25', '1.26']
 
# # Text on the top of each bar
for i in range(len(r4)):
    plt.text(x = r4[i] - 1.05 , y = data1[i] + 0.01, s = label1[i], size = 12)
    plt.text(x = r4[i]  - 0.85 , y = data2[i] + 0.01, s = label2[i], size = 12)


plt.xlabel("Classification model", fontsize=18)
plt.savefig('MI.png',bbox_inches='tight')
plt.show()
```
